# Load tikzmagic extension

````python
%load_ext tikzmagic
````

No special TeX commands for adding a node to a picture.

# First example: triangle

````python
%%tikz -f svg
\fill [fill=yellow!80!black]
    (0,0) node {First node}
--  (1,1) node [behind path] {Second node}
--  (2,0) node {Third node};
````

Syntax for specifying nodes:

```
\path ... node <foreach statements> [<options>] (<name>) at (<coordinate>) {<node contents>} ... ;
```

````python
%%tikz -f svg
\draw node[color=blue] (mynode) at (0,0) {1st Text in the node};
\draw node[color=red] (mynode) at (1.0,0.5) {2nd Text in the node};
````

````python
%%tikz -f svg -sc 0.5
\path (0,0) node [red] {A}
      (1,0) node [blue] {B}
      (2,0) node [green, node contents=C]
      (3,0) node [node contents=D];
````

````python
%%tikz -f svg -sc 0.5
\draw (0,0) node [red] {A} --
      (1,0) node [blue] {B}
      (2,0) node [green, node contents=C]
      (3,0) node [node contents=D];
````

# Location: behind or in front of the path

````python
%%tikz -f svg

\fill [fill=blue!50, draw=blue, very thick]
  (0,0) node [behind path, fill=red!50] {First node}
-- (1.5,0) node [behind path, fill=green!50] {Second node}
-- (1.5,1) node [behind path, fill=brown!50] {Third node}
-- (0,1) node [fill=blue] {Fourth node};
````

````python
%%tikz -f svg -p braket

\draw (0,0) -- (1,0) node[right] {$\ket{A}$};
\draw (0,1) -- (1,1) node[right] {$\ket{B}$};
````

````python
%%tikz -f svg -p braket

\draw (0,0) node[left] {$\ket{A}$} -- (1,0);
\draw (0,1) node[left] {$\ket{B}$} -- (1,1);
````

# Shape of a node

````python
%%tikz -f svg

\fill[fill=yellow!80!black]
   (0,0) node {First node}
-- (1,1) node [draw, behind path] {Second node}
-- (0,2) node [fill=red!20,draw,double,rounded corners] {Third node};
````

````python
%%tikz -f svg -l shapes.geometric

\fill[fill=yellow!80!black]
  (0,0) node {First node}
-- (1,1) node[ellipse,draw,behind path] {Second node}
-- (0,2) node[circle,fill=red!20] {Third node};
````

````python
%%tikz -f svg -p amsmath

\newcommand{\vecr}{\mathbf{r}}

\draw (0,0) node[name=SchEq,fill=blue!20]
{ $H \psi_{i}(\vecr) = \epsilon_{i} \psi_{i}(\vecr)$ };

\draw (0,-2) node[name=Hamiltonian,fill=blue!15]
{$ H = -\dfrac{1}{2}\nabla^2 + V_{\mathrm{KS}}(\vecr) $};

\draw[->] (node cs:name=SchEq,anchor=south) -- (0,-1) -- (-2.5,-1) |- (node cs:name=Hamiltonian,anchor=west);
````

````python
%%tikz -f svg -sc 1.5

\fill[gray!30] (0,0) -- (-2cm,0) -- (-2cm,3cm) -- (0,3cm) -- cycle;
\fill[gray!30] (2cm,0) -- (4cm,0) -- (4cm,3cm) -- (2cm,3cm) -- cycle;

\draw[->] (0,0) -- (0,3cm);
\draw (0,0) -- (2cm,0);
\draw[->] (2cm,0) -- (2cm,3cm);

\path node at (0cm,3.2cm) {$\infty$};
\path node at (2cm,3.2cm) {$\infty$};
\path node at (0cm,-0.5cm) {$0$};
\path node at (2cm,-0.5cm) {$a$};

\path node at (-1cm,1.5cm) {$\psi(x) = 0$};
\path node at (3cm,1.5cm) {$\psi(x) = 0$};

\draw[->] (0.75,-0.5) -- (1.25,-0.5) node[above,midway] {$x$};
````
