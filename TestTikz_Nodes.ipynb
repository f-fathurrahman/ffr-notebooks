No special TeX commands for adding a node to a picture.

# First example: triangle

````python
%%tikz -f svg
\fill [fill=yellow!80!black]
    (0,0) node {First node}
--  (1,1) node [behind path] {Second node}
--  (2,0) node {Third node};
````

Syntax for specifying nodes:

```
\path ... node <foreach statements> [<options>] (<name>) at (<coordinate>) {<node contents>} ... ;
```

````python
%%tikz -f svg
\draw node[color=blue] (mynode) at (0,0) {1st Text in the node};
\draw node[color=red] (mynode) at (1.0,0.5) {2nd Text in the node};
````

````python
%%tikz -f svg -sc 0.5
\path (0,0) node [red] {A}
      (1,0) node [blue] {B}
      (2,0) node [green, node contents=C]
      (3,0) node [node contents=D];
````

````python
%%tikz -f svg -sc 0.5
\draw (0,0) node [red] {A} --
      (1,0) node [blue] {B}
      (2,0) node [green, node contents=C]
      (3,0) node [node contents=D];
````
